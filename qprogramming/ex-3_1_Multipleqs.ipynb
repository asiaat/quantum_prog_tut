```python
## Programming Quantum Computers
##   by Eric Johnston, Nic Harrigan and Mercedes Gimeno-Segovia
##   O'Reilly Media
##
## More samples like this can be found at http://oreilly-qc.github.io

from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer, IBMQ, BasicAer
import math
## Uncomment the next line to see diagrams when running in a notebook
%matplotlib inline
```

```python
q1   = QuantumRegister(1, name='q1')
c1   = ClassicalRegister(1, name='c1')
q2   = QuantumRegister(1, name='q2')
c2   = ClassicalRegister(1, name='c2')
q3   = QuantumRegister(1, name='q3')
c3   = ClassicalRegister(1, name='c3')


qc    = QuantumCircuit(q1, c1,q2, c2,q3, c3)
```

```python
qc.reset(q1)

qc.reset(q2)
qc.h(q2)

qc.reset(q3)
qc.h(q3)
```

```python
qc.measure_all()
```

```python
qc.draw(output='mpl',justify='all')
```

```python
# Use Aer's qasm_simulator
simulator = Aer.get_backend('statevector_simulator')
```

```python
# Execute the circuit on the qasm simulator
job = execute(qc, simulator)

# Grab results from the job
result = job.result()

# Returns counts
counts = result.get_counts(qc)
print("\nTotal count for 00 and 11 are:",counts)
```
